Create a function to return whether a string is a palindrome. Also, create a fake dataset and filter it based on some conditions.

## Palindrome Script

```
def palindrome(input_str):
    # Convert any non-string inputs into string and lowercase
    forward_str = str(input_str).lower()

    # Reverse string to compare
    backward_str = forward_str[::-1]

    return forward_str == backward_str

print(f'Test Input: "hello"   - Results: {palindrome("hello")}   - Expected Results: False')
print(f'Test Input: "racecar" - Results: {palindrome("racecar")}    - Expected Results: True')
print(f'Test Input: "mom"     - Results: {palindrome("mom")}    - Expected Results: True')
print(f'Test Input: "DAd"     - Results: {palindrome("DAd")}    - Expected Results: True')
print(f'Test Input: 1337      - Results: {palindrome(1337)}   - Expected Results: False')
print(f'Test Input: 12.21     - Results: {palindrome(12.21)}    - Expected Results: True')
```
